### 掛け算
\begin{eqnarray}
\left[
\begin{array}{cc}
1 & 2 \\
3 & 4\\
\end{array}
\right]
\left[
\begin{array}{cc}
5 & 6 \\
7 & 8 \\
\end{array}
\right]
=
\left[
\begin{array}{ccccccc}
1x5+2x7 & 1x6+2x8 \\
3x5+4x7 & 3x6+4x8 \\
\end{array}
\right]
=
\left[
\begin{array}{cc}
19 & 22 \\
42 & 50 \\
\end{array}
\right]
\end{eqnarray}

#### ポイント：サイズが同じではない。(条件) M = O

行列A x B = C とした場合、
Aのサイズが(N x M), Bのサイズが(O x P)、 Cのサイズが(N x P)とした場合    
このとき、MのサイズとOのサイズが同じであれば、掛け算は成立する。  
また、CのサイズはNxPになる。  
ちなみに割り算は分数を使い掛け算にする $A / 2 = A * \frac{1}{2}$

### サイズ感（よく使う）
上記練習問題の1,2,3  

\begin{eqnarray}
\left[
\begin{array}{cccc}
ベ&ク&ト&ル \\
\end{array}
\right]
\left[
\begin{array}{c}
べ \\
ク \\
ト \\
ル \\
\end{array}
\right]
=
スカラー
\end{eqnarray}

横向きベクトルと縦向きのベクトルを掛けるとスカラーになる  
横向きベクトルは天地マークをつける　$x^T$
縦向きベクトルは何もつけない$y$  
つまりスカラーは「 $x^Ty$」と表す

\begin{eqnarray}
\left[
\begin{array}{c}
行列 \\
\end{array}
\right]
\left[
\begin{array}{c}
べ \\
ク \\
ト \\
ル \\
\end{array}
\right]
=
\left[
\begin{array}{c}
べ \\
ク \\
ト \\
ル \\
\end{array}
\right]
\end{eqnarray}

行列($X$)と縦向きベクトル($y$)を掛けると、縦向きベクトル「 $Xy$」になる  


\begin{eqnarray}
\left[
\begin{array}{cccc}
ベ&ク&ト&ル \\
\end{array}
\right]
\left[
\begin{array}{c}
行列 \\
\end{array}
\right]
\left[
\begin{array}{c}
べ \\
ク \\
ト \\
ル \\
\end{array}
\right]
=
スカラー
\end{eqnarray}

横向きベクトル($x^T$)、行列($A$)、縦向きベクトル($y$)なので、  
「$x^TAy$」 も実はスカラーを表す
