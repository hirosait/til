## ベクトルで微分

#### 例題１
\begin{eqnarray}
C=
\left[
\begin{array}{c}
3 \\
4 \\
\end{array}
\right],
X=
\left[
\begin{array}{c}
x_1 \\
x_2 \\
\end{array}
\right]
\end{eqnarray}
のとき  
$C^TX$  
は？
\begin{eqnarray}
=
\left[
\begin{array}{c}
3 & 4 \\
\end{array}
\right]
,
\left[
\begin{array}{c}
x_1 \\
x_2 \\
\end{array}
\right]
=
3x_1+4x_2
\end{eqnarray}


$C^TX$はスカラー  

#### 例題２



では、ベクトルで微分する場合の式を解いてみる  
  
$
\frac{\partial}{\partial X}(C^TX)
$  
  
  
$C^TX$は$(3x_1+4x_2)$
これを代入  


\begin{eqnarray}
=
\left[
\begin{array}{c}
\frac{\partial}{\partial X_1}(3x_1+4x_2) \\
\frac{\partial}{\partial X_2}(3x_1+4x_2) \\
\end{array}
\right]
\end{eqnarray}

上  
$\frac{\partial}{\partial X_1}(3x_1+4x_2)$  
$=\frac{\partial}{\partial X_1}(3x_1)+\frac{\partial}{\partial X_1}(4x_2)$  
$=3*\frac{\partial}{\partial X_1}(x_1)+4x_2*\frac{\partial}{\partial X_1}(1)$  
$=3*1+4x_2*0$  
$=3$

下  
$\frac{\partial}{\partial X_1}(3x_1+4x_2)$  
$=\frac{\partial}{\partial X_2}(3x_1)+\frac{\partial}{\partial X_2}(4x_2)$  
$=3*\frac{\partial}{\partial X_2}(1)+4*\frac{\partial}{\partial x_2}(x_2)$  
$=3x_1*0+4*1$  
$=4$  
  
つまり  
\begin{eqnarray}
=
\left[
\begin{array}{c}
3 \\
4 \\
\end{array}
\right]
\end{eqnarray}
  
これは転置が取れて元の$C$と同じになっている  
  
#### 例題２
$\frac{\partial}{\partial X}=$
\begin{eqnarray}
\left[
\begin{array}{c}
$\frac{\partial}{\partial X_1}(1) \\
$\frac{\partial}{\partial X_2}(1) \\
\end{array}
\right]
\end{eqnarray}
= 
\begin{eqnarray}
\left[
\begin{array}{c}
0 \\
0 \\
\end{array}
\right]
\end{eqnarray}
=
$\bf{0}$  
ゼロベクトルと呼ぶ
  
#### ベクトルで微分の公式
(1) $\frac{\partial}{\partial X}(a^Tx)=\bf{a}$

#### 例題３
$\frac{\partial}{\partial X}(C)$ Cは定数
= 
\begin{eqnarray}
\left[
\begin{array}{c}
$\frac{\partial}{\partial X_1}(c) \\
$\frac{\partial}{\partial X_2}(c) \\
\end{array}
\right]
\end{eqnarray}
=
\begin{eqnarray}
\left[
\begin{array}{c}
0 \\
0 \\
\end{array}
\right]
\end{eqnarray}

### ベクトルで微分の公式(重要)
(1) $\frac{\partial}{\partial C}=\bf{0}$  
(2) $\frac{\partial}{\partial C}(b^Tx)=\bf{b}$  
(3) $\frac{\partial}{\partial X}(x^TAx)=(A+A^T)\bf{x}$
