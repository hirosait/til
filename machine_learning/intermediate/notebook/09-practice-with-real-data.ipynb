# 入力変数と出力変数の切り分け

```python
# サーキットラーンにかけていく前に、最初に入力変数と出力変数の切り分けを行う
df.head(3)
```

# 訓練データと検証データを分割
- 例題：受験勉強用に10年分の過去問を購入してきたとする

だめなケース
- 10年分で勉強する（学習）
- 10年分で実力テスト（検証）

良いケース
- 前半の５年分で勉強する（学習）←これを訓練データ呼ぶ
- 後半の５年分で実力テスト（検証）←検証データと呼ぶ
訓練データと検証データを分ける。これも機械学習の一部なので、sklearnでできる。


```python
from sklearn.model_selection import train_test_split
```

```python
# 訓練データと検証データの分割
# 検証用データはtest_sizeで割合を指定。test_size=0.4は40%を検証、60%を訓練とする。
# 訓練データは前半5年分、後半5年分というわけかたではなく、ランダムに5年分ずつとなる。
# ランダムで取る場合、毎回ランダム値が変わると困るので、乱数シードをとっておき、再現性を確保しておく。random_state=1
# 結果はX_train, X_train, X_test, Y_train, y_test という順番で帰ってくる

X_train, X_test, y_train, y_test  = train_test_split(X,y, test_size=0.4, random_state=1)
```

```python
# モデルの学習（パラメータ調整）X_trainとY-trainを使う
model.fit(X_train, y_train)
```

```python
# 検証（訓練で使っていない検証データを使う）
model.score(X_test, y_test)
```

```python
# さっきより数値が悪いが、これが実際の値
# 検証する際には、検証データだけでなく、訓練データも見てみること
model.score(X_train, y_train)
```

```python
# ↑基本訓練データのほうが良い精度になる。なぜなら、訓練データに対してパラメータ調整をしているので、ベストな精度がでる。
# それに対して検証データは新しいデータなので、精度は劣化することが多いが、訓練データと検証データのスコアがそれほど離れていなければOK
# あとは、0.72とか0.72という数値が現実問題として、良い数値なのか悪い数値なのかの判断は業務を考え人間がする必要がある。
```
